```python
#EDA-Exploratory Data Analysis
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.hist(df['age'],bins=30,color='red',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Based EDA')
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

```python
# Income Distribution
plt.figure(figsize=(8,5))
df['income'].value_counts().plot(kind='bar',color=['violet','green'])
plt.xlabel('Income category')
plt.ylabel('Count')
plt.title('Income Distribution')
plt.grid(True,linestyle='--',axis='y')
plt.show()
```

```python
#Boxplot for Age vs Income
plt.figure(figsize=(8,5))
df.boxplot(column='age', by='income', grid=False)
plt.xlabel('Income category')
plt.ylabel('Age')
plt.title('Age Distribution by Income')
plt.suptitle("")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
for col in df.select_dtypes(include=['object']).columns:
        df[col]=LabelEncoder().fit_transform(df[col])
```

```python
# Split dataset
x=df.drop(columns=['income'])
y=df['income']
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Train linear Regression model

model=LinearRegression()
model.fit(X_train , y_train)
```

```python
# predict and evaluate
y_pred=model.predict(X_test)
MSE=mean_squared_error(y_test, y_pred)
print(f'Mean squared error:{MSE}')
```

```python
newPrediction_1=np.array([39,2,77777,4,0,13,2020,0,40,39,1,0,1,0])
```

```python
newPrediction_1=newPrediction_1.reshape(1,-1)
predictedIncome=model.predict(newPrediction_1)
```
